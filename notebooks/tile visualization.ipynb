```python
import os
import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from PIL import Image
from mgitools.os_helpers import listfiles
```

```python
st_fps = sorted(listfiles('/data/violet/st/pdac_ffpe_normalized/', regex=r'.jpeg'))
he_fps = sorted(listfiles('/data/tcia/PDA_preprocessed/', regex=r'.jpeg'))
```

```python
def show_tiles(fps, ncols=5, figsize=(5, 5)):
    nrows = (len(fps) // ncols)
    fig, axs = plt.subplots(nrows=nrows, ncols=ncols)
    queue = list(fps)
    for r in range(nrows):
        for c in range(ncols):
            im = Image.open(queue.pop())
            axs[r, c].imshow(np.asarray(im))
            axs[r, c].set_xticks([])
            axs[r, c].set_yticks([])
    plt.subplots_adjust(wspace=0, hspace=0)
```

```python
show_tiles(np.random.choice(st_fps, size=15))
```

```python
show_tiles(np.random.choice(he_fps, size=15))
```
