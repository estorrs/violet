```python
import os
import subprocess
from multiprocessing import Pool
import traceback

from mgitools.os_helpers import listfiles
```

```python
in_dir = '/data/tcia/PDA/'

out_dir = '/data/tcia/PDA_preprocessed_small_raw/'
script = '/home/estorrs/violet/violet/violet_preprocess.py'

ref = None
```

```python
fps = sorted(listfiles(in_dir, regex='.svs$'))
fps = [fp for fp in fps if '-21.svs' in fp]
len(fps), fps[:2]
```

```python
fps = fps[:20]
```

```python
# !rm -rf /data/tcia/PDA_preprocessed_small_raw/
```

```python
def get_svs_command(fp):
    new = fp.split('/')[-1]
    new_out = os.path.join(out_dir, new.split('.')[0])
    new = os.path.join(in_dir, new)
    if ref is not None:
        return f'python {script} svs {new} {new_out} --reference {ref}'
    else:
        return f'python {script} svs {new} {new_out}'
```

```python
commands = [get_svs_command(fp) for fp in fps]
commands[:2]
```

```python
def preprocess(fp):
    print(fp)
    command = get_svs_command(fp)
    try:
        out = subprocess.check_output(command, shell=True)
    except:
        print(f'failed: {fp}')
        traceback.print_exc()
    return fp
```

```python
# preprocess(fps[0])
```

```python
threads = 4
with Pool(threads) as p:
    print(p.map(preprocess, fps))
```
