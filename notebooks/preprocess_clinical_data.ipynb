```python
import pandas as pd
```

###### CCRCC

```python
df = pd.read_csv('/data/tcia/clinical/CCRCC_Apr2021_case.tsv', index_col='case_id', sep='\t')
df
```

```python
df = df[['follow-up/vital_status_at_date_of_last_contact', 'follow-up/days_from_date_of_initial_pathologic_diagnosis_to_date_of_last_contact']]
df = df.dropna()
df
```

```python
df['status_verbose'] = [x.split('|')[-1] for x in df['follow-up/vital_status_at_date_of_last_contact']]
df['status'] = [1 if x=='Living' else 2 for x in df['status_verbose']]
df['survival_time'] = [float(x.split('|')[-1]) if isinstance(x, str) else x
                       for x in df['follow-up/days_from_date_of_initial_pathologic_diagnosis_to_date_of_last_contact']]
df
```

```python
df = df.iloc[:, 2:]
df
```

```python
df.to_csv('/data/tcia/clinical/CCRCC_cleaned.txt', sep='\t')
```
